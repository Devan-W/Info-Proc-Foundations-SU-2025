```python
## Working with a heap using an ordered list. Heaps allow you to find the
## largest and smallest elements in the list, in addition to writing to the
## list (push) and removing from the list (pop).

import heapq

## Creating an initial heap
h = [10, 20, 15, 30, 40]
heapq.heapify(h)

print(h)

## Appending an element to list.
heapq.heappush(h, 5)

## Append another element, remove the smallest element.
heapq.heappush(h, 7)
print("push", h)
heapq.heappop(h)
print("pop", h)

## return three smallest values from heap.
min = heapq.nsmallest(3, h)
print("3 smallest elements:", min)

## return four largest values from heap.
maxi = heapq.nlargest(4, h)
print("4 largest elements:", maxi)
```

```python
import struct

## Pack values {1,2} into binary format. 'hh' is a format string that returns
## two bytes for each integer.
var = struct.pack('hh', 1, 2)
print(var)

## Unpack binary values corresponding to {1,2} --> 'hh' in a 2-tuple.
tup = struct.unpack('hh', var)
print(tup)
```

```python
## Create a class called "dog" that allows you to specify any possible dog
## you encounter. Objects are instances of the dog class where the attributes
## are stored. Class should be expressive: attributes and instances should
## allow you to spcify the class.

class Dog:
    species1 = "Pit Bull Mix"  ## Class attributes
    species2 = "German Sheppard"

    def __init__(self, name, age):
        self.name = name  ## Instance attributes
        self.age = age

## Creating objects of the Dog class
dog1 = Dog("Kious", 10)
dog2 = Dog("Della", 6)

print(dog1.name)
print(dog1.age)
print(dog1.species1)
print(dog2.name)
print(dog2.age)
print(dog2.species2)
```

```python
## Stack: operates on the LIFO principle (Last In, First Out).

stack = []
stack.append(1)  ## Push: {1,2,3,4}
stack.append(2)
stack.append(3)
stack.append(4)
stack.pop()  ## Pop: {4}
print(stack)  ## Output: {1,2,3}
```

```python
import networkx as nx
import matplotlib.pyplot as plt

## Create a directed graph
graph = nx.DiGraph()

## Add nodes
graph.add_nodes_from([1, 2, 3, 4, 5])

## Add edges with weights
graph.add_edge(1, 2, weight=1)
graph.add_edge(2, 3, weight=0.5)
graph.add_edge(3, 4, weight=0.2)
graph.add_edge(2, 4, weight=0.2)
graph.add_edge(1, 3, weight=1)
graph.add_edge(1, 4, weight=0.5)
graph.add_edge(3, 5, weight=0.8)
graph.add_edge(2, 5, weight=1)

## Access nodes and edges
print("Nodes:", graph.nodes)
print("Edges:", graph.edges(data=True))

## Visualize the graph (using matplotlib)
nx.draw(graph, with_labels=True)
plt.show()
```

```python
## Create a Pandas DataFrame with lists of dictionaries + row and column indexes.
import pandas as pd

## Initialize lists data.
data = [{'a': 1, 'b': 2},
        {'a': 5, 'b': 10, 'c': 20}]

## With two column indices, values same
## as dictionary keys
df1 = pd.DataFrame(data, index=['first',
                                'second'],
                   columns=['a', 'b'])

# With two column indices with
# one index with other name
df2 = pd.DataFrame(data, index=['first',
                                'second'],
                   columns=['a', 'b1'])

## print for first data frame
print(df1, "\n")

## Print for second DataFrame.
print(df2)
```
